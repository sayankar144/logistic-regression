```python
import numpy as np 
from sklearn import datasets as ds
```

```python
iris = ds.load_iris()
iris.keys()
```

```python
X  = iris['data'][:,3:4]
y = (iris['target'] == 2).astype(np.int8)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X, y)
```

```python
log_reg.intercept_
```

```python
log_reg.coef_
```

```python
X_new = np.linspace(0, 3, 1000).reshape(1000, 1)
y_proba = log_reg.predict_proba(X_new)[:,1:]
y_pred = log_reg.predict(X_new)
y_proba 
```

```python
from matplotlib import pyplot as plt
%matplotlib 
plt.scatter(X_new, y_pred, color = "red")
plt.plot(X_new, y_proba, color = "blue")
```
